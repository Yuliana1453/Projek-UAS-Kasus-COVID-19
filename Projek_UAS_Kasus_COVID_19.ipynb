### **_STUDI KASUS 1_**

**1. Bagaimana distribusi kasus COVID-19 di seluruh dunia?**

```python
import numpy as np
import plotly.express as px

# Inisialisasi log_scale_vals
log_scale_vals = list(range(0,50,2))
scale_vals = [2**val for val in log_scale_vals]

Cumulative_cases_COVID = covid2[['Date_reported', 'Country', 'Cumulative_cases']].sort_values('Date_reported')
Cumulative_cases_COVID = Cumulative_cases_COVID[Cumulative_cases_COVID.Cumulative_cases > 0]
Cumulative_cases_COVID['log2(Cumulative_cases)'] = np.log2(Cumulative_cases_COVID['Cumulative_cases'])
Cumulative_cases_COVID['Date_reported'] = Cumulative_cases_COVID['Date_reported'].dt.strftime('%d/%m/%Y')

fig = px.choropleth(Cumulative_cases_COVID, locations="Country", locationmode='country names',
                    color="log2(Cumulative_cases)", hover_name="Country", hover_data=['Cumulative_cases'],
                    projection="natural earth", animation_frame="Date_reported",
                    title='<b>Total Distribusi Kasus COVID-19 di Seluruh Dunia Sepanjang Waktu</b>',
                    color_continuous_scale="reds",
                   )

fig.update_layout(coloraxis={"colorbar": {"title":"<b>Total Kasus</b><br>",
                                          "titleside":"top",
                                          "tickmode":"array",
                                          "tickvals":log_scale_vals,
                                          "ticktext":scale_vals}
                            }
                 )

fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 1000
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['duration'] = 500

fig.show()
```

Dari map plot bisa dilihat perkembangan jumlah kasus covid di setiap negara dari waktu ke waktu.

### **_STUDI KASUS 2_**

**2. Bagaimana perkembangan tren mingguan dari jumlah kasus COVID-19 di negara Indonesia dan Malaysia?**

```python
import plotly.graph_objects as go

def plot_stats(country):

    covid2_country = covid2[covid2.Country == country]
    covid2_country.set_index('Date_reported', inplace=True)

    if not all(covid2_country.New_cases.isna()):
        layout = go.Layout(
            yaxis={'range':[0, max(0, covid2_country.New_cases.max() * 1.1)],
                  'title':'Coronavirus Cases'},
            xaxis={'title':''},
        )

        fig = px.area(covid2_country, x=covid2_country.index, y="New_cases",
                      title=f"<b>Perkembangan Tren Mingguan Jumlah Kasus COVID-19 di {country}<br>Sejak {covid2_country.index[0].strftime('%d-%m-%Y')} hingga {covid2_country.index[-1].strftime('%d-%m-%Y')}</b>",
                      template='plotly_dark')

        fig.update_traces(line={'width':5})

        fig.update_layout(layout)
        fig.show()
```

```python
plot_stats('Indonesia')
```

```python
plot_stats('Malaysia')
```

### **_STUDI KASUS 3_**

**3. Bagaimana hubungan antara jumlah orang yang divaksinasi dengan kasus baru COVID-19 terbanyak di 25 negara terhadap tingkat efektivitas vaksinasi?**

Pertama, kita membuat file gabungan variabel Country, New_cases, dan TOTAL_VACCINATIONS dari data covid global dan data vaksinasi covid. Kita juga mencari jumlah dari variabel New_cases dan TOTAL_VACCINATIONS berdasarkan negara. Setelah variabel yang diperlukan sudah menjadi satu data frame, lalu kita visualisasikan.

```python
new_covid=covid2.groupby(["Country"])[["New_cases"]].sum().reset_index()
new_vaksin = vaksin2_drop.groupby("COUNTRY")["TOTAL_VACCINATIONS"].sum().reset_index()
df_merged = pd.merge(new_covid,new_vaksin, how='inner', left_on=['Country'], right_on=['COUNTRY'])
df_merged = df_merged[['COUNTRY', 'New_cases', 'TOTAL_VACCINATIONS']]
df_merged
```

```python
top_25_countries = (
    df_merged.nlargest(25, 'New_cases')
    .rename(columns={'COUNTRY': 'country', 'New_cases': 'New Cases', 'TOTAL_VACCINATIONS': 'Total Vaccinations'})
)
top_25_countries
```

```python
fig, ax1 = plt.subplots(figsize=(15, 8))

# Bar pertama untuk New Cases
bar1 = ax1.bar(top_25_countries['country'], top_25_countries['New Cases'], color='b', alpha=0.7, label='New Cases')

# Menggunakan sumbu kedua untuk Total Vaccinations
ax2 = ax1.twinx()
bar2 = ax2.bar(top_25_countries['country'], top_25_countries['Total Vaccinations'], color='r', alpha=0.7, label='Total Vaccinations')

# Menambahkan label dan judul
ax1.set_xlabel('Countries')
ax1.set_ylabel('New Cases', color='b')
ax2.set_ylabel('Total Vaccinations', color='r')
plt.title('New Cases and Total Vaccinations by Country')
ax1.legend(bar1, ['New Cases'], loc='upper left')
ax2.legend(bar2, ['Total Vaccinations'], loc='upper right')

# Menambahkan label sumbu x dengan rotasi 45 derajat
ax1.set_xticks(range(len(top_25_countries['country'])))
ax1.set_xticklabels(top_25_countries['country'], rotation=45, ha='right')

# Menampilkan chart
plt.show()
```

```python
# Pilih hanya kolom numerik
numeric_data = top_25_countries.select_dtypes(include=[np.number])

# Hitung matriks korelasi
correlation_matrix = numeric_data.corr()

# Visualisasikan heatmap korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()

```
